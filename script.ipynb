*Importando as bibliotecas*

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

db = pd.read_csv('data/casos_mensais_para.csv')
```

Média

```python
media = db['casos_notificados'].mean().round(2)
print(f"A média dos casos de dengue no periodo de 2022 a 2026 é: {media}")
```

Mediana

```python
mediana = db['casos_notificados'].median().round(2)
print(f"A mediana dos casos de dengue no periodo de 2022 a 2026 é: {mediana}")
```

Max

```python
maximo = db['casos_notificados'].max()
print(f"O maior número de casos de dengue no periodo de 2022 a 2026 é: {maximo}")
```

Min

```python
minimo = db['casos_notificados'].min()
print(f"O menor número de casos de dengue no periodo de 2022 a 2026 é: {minimo}")
```

Desvio

```python
desvio = db['casos_notificados'].std().round(2)
print(f"O desvio padrão de casos de dengue no periodo de 2022 a 2026 é: {desvio}")
```
